## Merkezi Limit Teoremi
### Merkezi Limit Teoremi bize, herhangi bir popülasyondan yeteri kadar fazla sayıda örnek seçersek, popülasyon dağılımının ne olduğu farketmeksizin, bu örneklerin ortalamalarının dağılımının normal dağılıma yaklaşacağını söyler. Burada bahsedilen örneklerin ayrı ayrı dağımları değildir. Seçilen bütün örneklerin ortalamalarının dağılımıdır. 

#### Biz de python kodları yazarak bu durulumu doğrulayan bir örnek yapacağız.

```python
#Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 0 ile 500 arasında rasgele sayılardan oluşan bir numpy array oluşturduk
p = np.random.rand(1000) * 500
```

Yukarıda 0 ile 500 arasındaki sayılardan oluşan 1000 elemanlı bir array oluşturduk aslında. Bu array üzerinde denemelerimizi gerçekleştirebiliriz 

```python
print("Popülasyon ortalaması: ", p.mean()) 
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
sns.barplot(x=list(range(0, p.shape[0])), y=p)
```

Yukarıda gördüğümüz gibi sayılarımız 0 ile 500 arasında değişiyor. Grafik üzerinde göstermemin sebebi bu sayıların grafiğinin normal dağılmadığını göstermektir. Fakat bizim rasgele seçeceğimiz örneklerin ortalamalarının grafiği göreceğiz ki normal dağılacak.

### Rastgele örneklerin popülasyon içinden seçilmesi

Biz aşağıda popülasyon yani array'imiz içinden gelen her biri 30 elemandan oluşan 30 tane ayrı örnek oluşturacağız. Bunu aşağıdaki gibi yapalım 

```python
x1 = []
for i in range(1):
    x1.append(np.random.choice(p, 30, replace=False))
x2 = []
for i in range(10):
    x2.append(np.random.choice(p, 30, replace=False))
x3 = []
for i in range(30):
    x3.append(np.random.choice(p, 30, replace=False))
x4 = []
for i in range(50):
    x4.append(np.random.choice(p, 30, replace=False))
x5 = []
for i in range(100):
    x5.append(np.random.choice(p, 30, replace=False))
x6 = []
for i in range(200):
    x6.append(np.random.choice(p, 30, replace=False))
```

Biz yukarıda p içerisinden 30 defa 30 elemanlı alt kümeler seçtik ve her örneği x listesinin içine ekledik. Şimdi x içindeki array'lerin her birinin ortalamasının grafiğini çizdirelim.

```python
means1 = [i.mean() for i in x1]
means2 = [i.mean() for i in x2]
means3 = [i.mean() for i in x3]
means4 = [i.mean() for i in x4]
means5 = [i.mean() for i in x5]
means6 = [i.mean() for i in x6]
sns.histplot(x=means1)
plt.title("1 Örnek Seçildiğinde")
plt.show()
sns.histplot(x=means2)
plt.title("10 Örnek Seçildiğinde")
plt.show()
sns.histplot(x=means3)
plt.title("30 Örnek Seçildiğinde")
plt.show()
sns.histplot(x=means4)
plt.title("50 Örnek Seçildiğinde")
plt.show()
sns.histplot(x=means5)
plt.title("100 Örnek Seçildiğinde")
plt.show()
sns.histplot(x=means6)
plt.title("200 Örnek Seçildiğinde")
plt.show()
```

Yukarıda gördük ki aslında bir çok örneğin ortalaması popülasyon ortalamasına çok yakın. Popülasyon ortalaması da en üstte yazdığımız gibi 250 civarında ki biz 0 ile 500 aralığında rasgele sayılar seçtiğimiz için bu çok doğal. Fakat bazı örnekler ortalamadan uzaklaşmış ve ortalamadan uzak olanların sayısı da ortalamaya yakın olanlara göre gittikçe azalıyor. Yani seçtiğimiz örneklerin ortalaması genel olarak popülasyon ortalaması etrafında dağılmaktadır. Hatta bu dağılım normal bir dağılım gösteriyor.

Yukarıda ne kadar çok örnek seçersek örneklerin ortalamaları normal dağılıma o kadar çok yaklaşıyor. Dolayısıyla en başta söylediğimiz tanımı burada yenileyerek bitirebiliriz.

### Merkezi Limit Teoremi

Bir popülasyondan seçilen örneklerin ortalamaları popülasyon ortalamasının dağılımından bağımsız olarak normal dağılım gösterir. Örnek ortalamalarının normal dağılım göstermesi için yukarıda gördüğümüz gibi seçilen örnek sayısının yeterli olması gerekmektedir. Yani sadece bir veya iki örnek seçtiğimizde ortalamaları popülasyon ortalamasına göre normal dağılmasa bile, örnek sayısı arttıkça örneklerin ortalamalarının çoğu popülasyon ortalaması etrafında toplanarak normal dağılım oluşturur. Yoksa 1 tane veya 10 tane örnek seçmek yukarıdaki grafiklerde görüldüğü gibi yeterli değil.

İşte bu sayede biz bir anket yapmak istediğimizde en az 30 tane örnek seçip bu örneklerin ortalamasının popülasyon ortalamasını temsil ettiğini ve yakın olduğunu kabul edebiliriz. Dolayısıyla da anketlerdeki örneklerin popülasyondaki ile benzer özelliklere sahip olduğunu kabul edip çıkarım yapabiiliriz.
