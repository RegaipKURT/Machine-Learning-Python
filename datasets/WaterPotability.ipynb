```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("water_potability.csv")
print(data.shape)
```

```python
sns.pairplot(data)
```

```python
print("Column Informations:")
data.info()
print("-"*50)
print("Null Values Count:")
print(data.isnull().sum())
print("-"*50)
print("Null Values Count Rate:")
print(data.isnull().sum() / data.shape[0])
```

```python
data.Potability = data.Potability.astype("category")
data.ph.isna()
```

```python
nans = ["ph", "Sulfate", "Trihalomethanes"]

for nan_col in nans:
    data[nan_col + "_na"] = data[nan_col].isna()
    data[nan_col].fillna(value=data[nan_col].mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x, y = data.drop(labels="Potability", axis=1), data["Potability"]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
```

```python
print(X_test.shape, X_train.shape)
print(y_test.shape, y_train.shape)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, bootstrap=True, max_features=0.5, n_jobs=-1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("F1 Score: ", f1)
print("Accuracy Score: ", ac)
print("CM: \n", cm)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_scaled = sc.fit_transform(x)

x = np.array(x, dtype=np.float64)
y = np.array(y, dtype=np.number)

xs = x_scaled
XS_train, XS_test, ys_train, ys_test = train_test_split(xs, y, test_size=0.1)

model = RandomForestClassifier(n_estimators=50, bootstrap=True, max_features=0.5, n_jobs=-1)
model.fit(XS_train, ys_train)
model.score(XS_test, ys_test)

ys_pred = model.predict(XS_test)

cm = confusion_matrix(ys_test, ys_pred)
ac = accuracy_score(ys_test, ys_pred)
f1 = f1_score(ys_test, ys_pred)

print("F1 Score: ", f1)
print("Accuracy Score: ", ac)
print("CM: \n", cm)
```

```python
from keras.models import Sequential
from keras.layers import Dense


model = Sequential()
model.add(Dense(128, activation="relu"))
model.add(Dense(256, activation="tanh"))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="mse", metrics=["acc"])
model.fit(XS_train, ys_train, epochs=500, verbose=0)

ys_pred = model.predict(XS_test)

ys_pred = np.array(ys_pred > 0.5, dtype=np.int)

cm = confusion_matrix(ys_test, ys_pred)
ac = accuracy_score(ys_test, ys_pred)
f1 = f1_score(ys_test, ys_pred)

print("F1 Score: ", f1)
print("Accuracy Score: ", ac)
print("CM: \n", cm)
```
